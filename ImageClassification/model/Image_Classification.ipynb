### Let's try to detect the face and the eyes. OpenCV contains Haar Cascades algorithm which is an Object Detection Algorithm used to identify faces in an image or a real time video. It is a machine learning based approach where a cascade function is trained from a lot of positive(images of faces) and negative images(images without faces). It is then used to detect objects in other images.They allow you to detect different features on the face and contains pre-trained classifiers. 

```python
face_cascade = cv2.CascadeClassifier('./opencv/haarcascades/haarcascade_frontalface_default.xml')
eye_cascade  = cv2.CascadeClassifier('./opencv/haarcascades/haarcascade_eye.xml')

faces = face_cascade.detectMultiScale(gray, 1.3, 5)
faces
```

**(352,  38, 233, 233) are x,y,width,height**

```python
(x,y,w,h) = faces[0]
x,y,w,h
```

**Drawing a rectangle around the face**

```python
face_img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
#face_img = cv2.rectangle(img,(x1,y1),(x2,y2),(R,G,B),width of line)
plt.imshow(face_img)
```

**Drawing rectangles on 2 eyes and the face. Steps:** <br>
**1. Detect a Face using Face cascade and draw a rectangle around it.** <br>
**2. Isolate the face window and divide it into Gray and Color.** <br>
**3. Detect eyes using eyes cascade on isolated Gray Window of the Face.**  <br>
**4. Draw rectangles on the eyes on the isolated Color Window and since it's a subset of Face window, changes will take place overall on Face window**

```python
cv2.destroyAllWindows()
for (x,y,w,h) in faces:
    face_img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = face_img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        print (ex,ey,ew,eh)
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),3)
        

plt.figure()
plt.imshow(face_img, cmap='gray')
plt.show()
#Region of interest: ROI
```

### We want these kind of facial regions for our classifier model. So we would like to detect these faces in all images, crop those images and use them later. Let's write a function where we can input an image and it returns the cropped face if the face and two eyes are detected clearly.

```python
def get_cropped_image(image_path):
    img = cv2.imread(image_path) # Loading Image
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # Getting Grayscale Image
    faces = face_cascade.detectMultiScale(gray, 1.3, 5) # Getting Faces
    for (x,y,w,h) in faces: # Iterating through Faces
        roi_gray = gray[y:y+h, x:x+w] # Separating ROI in gray
        roi_color = img[y:y+h, x:x+w] # Separating ROI in color
        eyes = eye_cascade.detectMultiScale(roi_gray) # Detecting eyes on ROI: gray 
        if len(eyes) >= 2: # Min number of eyes required for the return
            return roi_color
```

# Section 2 : Preparing Train Data using Wavelet Transform

**Now we are going to implement Feature engineering. We will be using a techinque called wavelet transform which is used to extract the facial features. We will build our training datasets using this transform for the model.**

### Wavelet Transform

```python
# Code taken from stackoverflow
import pywt
def w2d(img, mode='haar', level=1):
    imArray = img
    #Datatype conversions
    #convert to grayscale
    imArray = cv2.cvtColor( imArray,cv2.COLOR_RGB2GRAY )
    #convert to float
    imArray =  np.float32(imArray)   
    imArray /= 255;
    # compute coefficients 
    coeffs=pywt.wavedec2(imArray, mode, level=level)

    #Process Coefficients
    coeffs_H=list(coeffs)  
    coeffs_H[0] *= 0;  

    # reconstruction
    imArray_H=pywt.waverec2(coeffs_H, mode);
    imArray_H *= 255;
    imArray_H =  np.uint8(imArray_H)

    return imArray_H
```

### Preparing Training Data

```python
X = []
y = []

for sportsperson, all_files in Sportsperson_file_names_Improved.items():
    for training_image in all_files:
        img = cv2.imread(training_image)
        scaled_raw_img = cv2.resize(img, (32, 32))
        image_wt = w2d(img,'db1',5)
        scaled_img_wt = cv2.resize(image_wt, (32, 32))
        combined_img = np.vstack((scaled_raw_img.reshape(32*32*3,1),scaled_img_wt.reshape(32*32,1)))
        X.append(combined_img)
        y.append(class_dict[sportsperson])
```

# Section 3: Building the Model & Hypertuning

**We now have our Training Data (X,y). We are going to build & train the model, do hyperparameter tuning using GridsearchCV and then we will select the best model for our image classification. We will start with Support Vector Machine(SVM), because SVM tends to perform good with image calssification. We will try a couple of other models using GridsearchCV. The goal is to observe the perfomance of different models using gridsearchCV, do hyperparameter tuning and in the end select the model with the optimum performance.**

## Let's start with building the model using SVM

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
#Splitting Data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=101)

# Creating Sklearn pipeline
# First we scale our Data and then we select SVM classifier to perform on the data
# Pipeline(steps, *, memory=None, verbose=False)
pipe = Pipeline([('scaler', StandardScaler()), ('svc', SVC(kernel = 'rbf', C = 10))],verbose=2)
# Using Pipeline we sequentially apply a list of transforms and a final estimator.
# pipe can be used as any other estimator

pipe.fit(X_train, y_train)
pipe.score(X_test, y_test)
```

### Now we use GridSearchCV to hypertune parameters. We use it with different classifiers as mentioned in the following cells. GridSearchCV informes about the models that performed best and the set of parameters that worked well for the model. Ultimately, it gives us the best estimator for a specific classifier.

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
```

### Classifiers selected:
**1. Support Vector Classifier** <br>
**2. Random Forest** <br>
**3. Logistic Regression**
---


**Let's define different classifiers as mentioned above. We want to try these classifiers with different specific parameters to select the best estimator for a particular classifier.**

```python
model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto',probability=True),
        'params' : {
            'svc__C': [1,10,100,1000],
            'svc__kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'randomforestclassifier__n_estimators': [1,5,10]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'logisticregression__C': [1,5,10]
        }
    }
}
```

```python
for algo_name, algo in model_params.items():
    print (algo['model'])
    print (algo['params'])
```

**Now we iterate through the above dictionary, we select each model one by one and select the different parameters associated with that model. So for example, we first select the SVM model in the following manner : <br>
algo['model'] = SVC(gamma='auto', probability=True) and the associated parameter will be: <br>
algo['params'] = 'svc__C': [1, 10, 100, 1000], 'svc__kernel': ['rbf', 'linear'] <br>
We then repeat this process for the other models. We use the validation data within the training data to hypertune the parameters of the model. Later on to actually judge the performance of the model, we use the test data.**


```python
import pandas as pd

scores = []
best_estimators = {}
for algo_name, algo in model_params.items():
    pipe = make_pipeline(StandardScaler(), algo['model'])
    classifier =  GridSearchCV(pipe, algo['params'], cv=5, return_train_score=False,verbose=2)
    classifier.fit(X_train, y_train)
    scores.append({
        'model': algo_name,
        'best_score': classifier.best_score_,
        'best_params': classifier.best_params_
    })
    print (scores)
    best_estimators[algo_name] = classifier.best_estimator_
    
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
```
